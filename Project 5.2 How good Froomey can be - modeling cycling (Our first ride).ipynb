## 1. Importing our team

Let me remind previous notebook:

```python
global g
dt = 1                        #time step
g = 9.8                       #gravitational constant
order_coeff = np.log(2)/4    #slipstream constant     
bar_quant = 100               #number of bar_quants [J]
bar_full = 300/dt * bar_quant #capacity of bar
rho = 1.225                   #density of air
mu = 0.001                    #rolling friction coefficient
basic_air_coeff = 0.3

class Cyclist():
    
    def __init__(self,energy, mass, area, bars, name,order):
        '''
            Class cyclists with fields 
                - Energy : reservoir of cyclists energy
                - Area : wind area of cyclists
                - Mass : 
                - Order: place in peleton
                - Bars : number of energy bars
                - Speed: Default = 0
                - Distance: Default = 0
                - Name : Random name
                - Data: List of cyclits values during simulation
                - Other variables are not so important
                
        '''
        self.energy = np.round(energy,2)
        self.area = np.round(area,2)
        self.mass = np.round(mass,2)
        self.bars = bars
        self.order = order
        self.name = name
        self.speed = 0
        
        self.air_coef = basic_air_coeff
        self.distance = 0
        self.eated = False
        self.energy_added = 0
        self.maruder = False
        self.data = [[],[],[],[],[]]
    
    def __str__(self):
        return "Hi, my name is "+str(self.name)+" and my skills are as follows: \n Energy = "+str(self.energy)+"\n Mass = "+str(self.mass)+"\n Ares = "+str(self.area)+"\n I will be "+str(self.order)+"th in peleton \n\n"
    
    def set_velocity(self,v):
        '''
            Method setting cyclists speed
        '''
        self.speed = v
        
    def set_order(self,o):
        '''
            Method setting cyclists order in peleton
        '''
        self.order = o
    
    def use_slipstream(self):
        '''
            Calculating changed air_coef using simple model
        '''
        self.air_coef = basic_air_coeff * np.exp(-order_coeff * self.order)
        
    def drink_water(self):
        #to be implemented
        return null
        
    def eat_energy_bar(self):
        if self.bars > 0:
            self.eated = True
            self.bars -= 1
        
#-----------------------------------------------------------------------------------------------------------
#Calculation methods:
    def calc_air_drag(self, rho):
        return self.air_coef * self.area * rho * self.speed**2
    
    def calc_rolling_drag(self, mu):
        return self.mass * g * mu
    
    def calc_incline_drag(self, alpha):
        return self.mass * g * alpha
    
    def calc_power(self,rho, mu, alpha):
        return (self.calc_air_drag(rho) + self.calc_rolling_drag(mu) + self.calc_incline_drag(alpha)) * self.speed
    
#-----------------------------------------------------------------------------------------------------------
    def update(self,rho, mu, alpha, dt):
        '''
            Method updating status of cyclits in each iteration
        '''
        if self.eated:
            self.energy_added += bar_quant * dt
            self.energy += bar_quant * dt
            
            if self.energy_added == bar_full:
                self.eated = False
                self.energy_added = 0
        
        self.use_slipstream()
        self.energy -= self.calc_power(rho, mu, alpha) * dt
        self.distance += self.speed * dt
        self.data[0].append(self.distance)
        self.data[1].append(self.speed)
        self.data[2].append(self.energy)
        self.data[3].append(self.order)
        self.data[4].append(self.calc_power(rho, mu, alpha))    

cyclists = []
Names = []

min_energy = 0.8 * 1e6 #[J]
min_mass = 60        #[kg]
min_area = 0.3       #[m^2]
random_factor = 0.2  #possible addon to basic number
bars = 1

#initialization
for i in range(9):
    energy = min_energy * (1 + random_factor*random.random())
    mass = min_mass * (1 + random_factor*random.random())
    area = min_area * (1 + random_factor*random.random())
    bars = 1
    name = names.get_full_name(gender='male')
    cyclists.append(Cyclist(energy, mass, area, bars, name , 0))

#Best strategy seems to have cyclists ordered in increasing energy value:
cyclists.sort(key=lambda x: x.energy, reverse=False)

for i,cycl in enumerate(cyclists):
    cycl.set_order(i)
    print(cycl)
```

## 2. Selecting the terain:

Those numbers are the gradients, which mathematically speaking are equal to $tan\alpha$, which means to be over precise I would like to do following transformation

```python
mount_ventoux_slope = np.sin(np.arctan(np.array([2.6, 3.6, 4.4, 6.0, 5.9, 4.5, 9.7, 9.6, 10.9, 9.2, 
                      10.1, 9.7, 8.7, 10.4, 8.1, 7.6, 5.9, 7.1, 7.4, 7.9, 9.9]) * 0.01))
flat_slope = np.array([0.0]*21)

#select slope:
global slope
#slope = mount_ventoux_slope
slope = flat_slope
```

## 3. The race:


The race is controlled by deterministic parameters: bar_plan, speed_plan responsible for taking energy_bars and speed_plan for each km respectively. In first case I would like to assume constant speed and random bar_plan:

The race class looks as follows:

```python
def The_race(bars_plan, speed_plan):
    '''
        Method simulating 21 km race
    '''
    global cyclist_team
    #initialization of variable
    t = 0
    maruders = 0
    change = False

    while cyclist_team[0].distance < 21 * 1e3:
        ''' 
            While first guy covers distance less than 21 km
        '''
        t += dt #time update
        
        ind = int(np.floor(cyclist_team[0].distance/1e3))  #km counter
        
        for i, cycler in enumerate(cyclist_team):
            if not cycler.maruder: #is a rider in a pack? If yes: update
                
                cycler.set_velocity(speed_plan[ind])
                cycler.update(rho, mu, slope[ind], dt)
                if change:
                    cycler.order -= 1
                
                if bars_plan[i] == ind: #eat bar
                    cycler.eat_energy_bar()

                if cycler.energy < 0: #is rider exhausted?
                    cycler.set_order(len(cyclist_team) - 1 - maruders) #take exhausted rider to the end
                    cycler.set_velocity(1)  #travel with grupetto speed
                    cycler.maruder = True   #be a maruder
                    change = True
                    maruders +=1
            else:
                cycler.update(rho, mu, mount_ventoux_slope[ind], dt) #if already in grupetto just update
        #sort list with order key
        cyclist_team.sort(key=lambda x: x.order, reverse=False)
        change = False
    print(t)
    return t
        
        
```

```python
global cyclist_team
bars_plan = np.random.randint(0, 21, size= 9)  #on wich km should rider take energy_bar
constant_speed = 65   #conatnt speed [km/h]
speed_plan = 1/3.6 *np.array([constant_speed]* 21)
cyclist_team = copy.deepcopy(cyclists)
t = The_race(bars_plan, speed_plan)
```
